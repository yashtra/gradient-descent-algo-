```python
#creating a gradient descent algo from scratch
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#  a linear regression model
def linear_model(w,b,i):
  y=[]
  for x in range(i):
    y.append(w*x+b)

  return y

def sum_errb(tw,tb,w,b,i):
  sum_errors=0
  t_y=linear_model(tw,tb,i)
  g_y=linear_model(w,b,i)
  for r in range(i):
   sum_errors+=(g_y[r]-t_y[r])
  return sum_errors

def sum_errw(tw,tb,w,b,i):
  x=list(range(i))
  sum_errors=0
  t_y=linear_model(tw,tb,i)
  g_y=linear_model(w,b,i)
  for r in range(i):
   sum_errors+=(g_y[r]-t_y[r])*x[r]

  return sum_errors
```

```python
def grd_dscnt(tw,tb,sw,sb,i,a):
 tar_y=linear_model(tw,tb,i)
 x=list(range(i))
 plt.scatter(x,tar_y)
 #gradient descent algorithm
 w=sw
 b=sb
 for r in range(1100):

  sum_errorb=sum_errb(tw,tb,w,b,i)
  sum_errorw=sum_errw(tw,tb,w,b,i)
  print(sum_errorw,'w:' ,w )
  temp_w=w-((a/(i-1))*sum_errorw)
  temp_b=b-((a/i)*sum_errorb)
  w=temp_w
  b=temp_b
  pred_y=linear_model(w,b,i)
  plt.plot(x,pred_y,color='green')

 pred_y=linear_model(w,b,i)
 plt.plot(x,pred_y,color='red')







grd_dscnt(2,4,5,3,5,0.1)

```
